```python
import numpy as np
```

```python
N = 1000
data = np.zeros(N//8, dtype = np.dtype("u1"))
for i in range(N//8):
    st = ""
    for k in range(8):
        st += "{}".format(np.random.choice([0,1]))
        num = int(st, 2)
        data[i] = num
```
